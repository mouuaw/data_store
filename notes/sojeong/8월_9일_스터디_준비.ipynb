```
df_tr = np.array(train.iloc[:,1:], dtype = 'float32')
df_ts = np.array(test.iloc[:,1:], dtype='float32')
```

```
x_train = df_tr[:,1:]/255
y_train = df_tr[:,0]
x_test= df_ts/255
```

```
x_train,x_valid,y_train,y_valid = train_test_split(x_train,y_train,test_size = 0.2,random_state = 12345)
```

```
x_train = x_train.reshape(x_train.shape[0],28,28,1)
x_test = x_test.reshape(x_test.shape[0],28,28,1)
x_validate = x_validate.reshape(x_validate.shape[0],28,28,1)
```

```
def solution_model():
    model = Sequential([
        # 1번
        # Flatten(input_shape = (28,28)),
        # Reshape((28,28,1)),
        # Conv2D(16,(3,3),padding='same',activation='relu'),
        # MaxPooling2D(pool_size=(2,2),padding='same'),
        # Conv2D(32,(3,3),padding='same',activation='relu'),
        # MaxPooling2D(pool_size=(2,2),padding='same'),        
        # Conv2D(64,(3,3),padding='same',activation='relu'),
        # MaxPooling2D(pool_size=(2,2),padding='same'),            
        # Flatten(),
        # Dense(512, activation = 'relu'),
        # Dense(256, activation = 'relu'),
        # Dense(64, activation = 'relu'),
        # Dropout(0.2),
        # Dense(10, activation = 'softmax')
        # 2번
        # Flatten(input_shape=(28, 28)),
        # Dense(1024, activation='relu'),
        # Dense(512, activation='relu'),
        # Dense(256, activation='relu'),
        # Dense(128, activation='relu'),
        # Dense(64, activation='relu'),
        # Dense(10, activation='softmax')
        #3번
        Flatten(input_shape = (28,28)),
        Reshape((28,28,1)),
        Conv2D(32,(3,3),padding='same',activation='relu'),
        MaxPooling2D(pool_size=(2,2),padding='same'),
        Conv2D(32,(3,3),padding='same',activation='relu'),
        MaxPooling2D(pool_size=(2,2),padding='same'),        
        Conv2D(64,(3,3),padding='same',activation='relu'),
        MaxPooling2D(pool_size=(2,2),padding='same'),            
        Flatten(),
        Dense(512, activation = 'relu'),
        Dense(256, activation = 'relu'),
        Dense(128, activation = 'relu'),
        Dense(64, activation = 'relu'),
        Dropout(0.2),
        Dense(10, activation = 'softmax')
    ])
   
    #Compile
    model.compile(optimizer='Adam',loss='sparse_categorical_crossentropy',metrics = ['acc'])
   
    lowest_point = "tf_model.ckpt"
    lowest_loss = ModelCheckpoint(filepath=lowest_point,
                             save_weights_only=True,
                             save_best_only=True,
                             monitor='val_loss',
                             verbose=1)
   
    final_model = model.fit(x_train, y_train,
                    validation_data=(x_valid, y_valid),
                    epochs=10,
                    callbacks=[lowest_loss],
                   )
   
    model.load_weights(lowest_point)

    return model
```

```
if __name__ == '__main__':
    model = solution_model()
    model.save("mymodel.h5")
```
